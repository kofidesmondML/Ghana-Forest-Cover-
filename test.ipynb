```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import Input

def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data from {file_path}: {e}")
        return None

def prepare_train_test_data(district_data, sequence_length=10):
    try:
        losses = district_data.loc[:, 'tc_loss_ha_2001':'tc_loss_ha_2023'].values.flatten()
        if len(losses) <= sequence_length:
            print(f"Not enough data for sequence preparation in district: {district_data['subnational2'].iloc[0]}")
            return None, None, None, None
        
        scaler = MinMaxScaler()
        losses_scaled = scaler.fit_transform(losses.reshape(-1, 1)).flatten()
        
        X, y = [], []
        for i in range(len(losses_scaled) - sequence_length):
            X.append(losses_scaled[i:i + sequence_length])
            y.append(losses_scaled[i + sequence_length])
        
        X, y = np.array(X).reshape((-1, sequence_length, 1)), np.array(y)
        
        # Split into train/test (2001-2018 for training, 2019-2023 for testing)
        train_size = 18  # 2001–2018
        X_train, y_train = X[:train_size], y[:train_size]
        test_scaled = losses_scaled[train_size:]
        
        return X_train, y_train, test_scaled, scaler
    except Exception as e:
        print(f"Error preparing data: {e}")
        return None, None, None, None

def build_lstm_model(input_shape):
    try:
        model = Sequential([
            Input(shape=input_shape),
            LSTM(50, activation='relu'),
            Dropout(0.2),
            Dense(1)
        ])
        model.compile(optimizer='adam', loss='mean_squared_error')
        return model
    except Exception as e:
        print(f"Error building LSTM model: {e}")
        return None

def evaluate_predictions(test_scaled, predictions, scaler):
    try:
        test_actual = scaler.inverse_transform(test_scaled.reshape(-1, 1)).flatten()
        test_predicted = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()
        mae = mean_absolute_error(test_actual, test_predicted)
        mse = mean_squared_error(test_actual, test_predicted)
        r2 = r2_score(test_actual, test_predicted)
        return mae, mse, r2, test_actual, test_predicted
    except Exception as e:
        print(f"Error evaluating predictions: {e}")
        return None, None, None, None, None

def plot_all_years(district_name, train_years, train_values, test_years, test_actual, pred_years, predictions):
    train_values = train_values[-len(train_years):]  # Align values to match years if necessary
    plt.figure(figsize=(12, 6))
    plt.plot(train_years, train_values, label='Training Data (2001-2018)', color='blue', marker='o')
    plt.plot(test_years, test_actual, label='Actual Data (2019-2023)', color='green', marker='o')
    plt.plot(pred_years, predictions, label='Predictions (2019-2023)', color='red', linestyle='dashed', marker='o')
    plt.title(f"Prediction Results for {district_name}")
    plt.xlabel("Year")
    plt.ylabel("Forest Loss")
    plt.legend()
    plt.grid()
    plt.savefig(f"./plots/all_years_{district_name}.png")

def process_district(district_data):
    try:
        if 'subnational2' not in district_data.columns or district_data.empty:
            raise ValueError("Invalid or empty district data")
        
        district_name = district_data['subnational2'].iloc[0]
        print(district_name)
        
        X_train, y_train, test_scaled, scaler = prepare_train_test_data(district_data)
        if X_train is None or y_train is None or len(X_train) == 0 or len(y_train) == 0:
            raise ValueError("Training data is missing or empty")
        
        if len(X_train.shape) != 3:
            raise ValueError("X_train should be a 3D array")
        
        model = build_lstm_model((X_train.shape[1], 1))
        
        try:
            model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0)
        except Exception as fit_error:
            raise ValueError(f"Error during model training: {fit_error}")
        
        if X_train.shape[0] == 0:
            raise ValueError("X_train is empty, cannot generate last sequence")
        
        last_sequence = X_train[-1].reshape(1, X_train.shape[1], 1)
        predictions = []
        for _ in range(len(test_scaled)):
            next_pred = model.predict(last_sequence, verbose=0)
            predictions.append(next_pred[0, 0])
            last_sequence = np.append(last_sequence[:, 1:, :], next_pred.reshape(1, 1, 1), axis=1)
        
        mae, mse, r2, test_actual, test_predicted = evaluate_predictions(test_scaled, predictions, scaler)
        print(f"District: {district_name} | MAE: {mae:.4f}, MSE: {mse:.4f}, R2: {r2:.4f}")
        
        train_years = list(range(2001, 2019))
        test_years = list(range(2019, 2024))
        pred_years = list(range(2019, 2024))
        train_values = scaler.inverse_transform(X_train[:, -1, 0].reshape(-1, 1)).flatten()
        print(f'train values are {len(train_values)}')
        
        try:
            plot_all_years(district_name, train_years, train_values, test_years, test_actual, pred_years, test_predicted)
        except Exception as plot_error:
            raise ValueError(f"Error during plotting: {plot_error}")
        
        return test_predicted
    except Exception as e:
        import traceback
        print(f"Error processing district: {e}")
        print(traceback.format_exc())
        return None


def main():
    forest_path = './clean_data/forest.csv'
    forest_df = load_data(forest_path)
    if forest_df is None:
        return

    predictions_dict = {}
    districts = forest_df['subnational2'].unique()

    for district in districts:
        try:
            district_data = forest_df[forest_df['subnational2'] == district]
            #print(district_data)
            print('checking here')
            predictions = process_district(district_data)
            print('checking done')
            print(predictions)
            if predictions is not None:
                predictions_dict[district] = predictions
        except Exception as e:
            print(f"Error processing district {district}: {e}")
    
    predictions_df = pd.DataFrame(predictions_dict)
    predictions_df.to_csv('./clean_data/lstm_predictions.csv', index=False)
    print("Predictions saved successfully.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np

forest_df = pd.read_csv('./clean_data/forest.csv')
tc_loss_columns = [col for col in forest_df.columns if 'tc_loss_ha_' in col]
tc_loss_data = forest_df[tc_loss_columns].mean(axis=0)
tc_loss_data.index = range(2001, 2024)

tc_loss_data.plot(figsize=(10, 6))
plt.title('Tree Cover Loss (Ha) Over Time')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.show()

result = adfuller(tc_loss_data)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] > 0.05:
    print("The time series is non-stationary. Differencing is required.")
else:
    print("The time series is stationary.")

if result[1] > 0.05:
    tc_loss_data_diff = tc_loss_data.diff().dropna()
else:
    tc_loss_data_diff = tc_loss_data

plot_acf(tc_loss_data_diff)
plt.show()
plot_pacf(tc_loss_data_diff)
plt.show()

best_aic = np.inf
best_order = None
best_model = None

for p in range(0, 4):
    for d in range(0, 2):
        for q in range(0, 4):
            try:
                model = ARIMA(tc_loss_data, order=(p, d, q))
                model_fit = model.fit()
                print(f"ARIMA({p}, {d}, {q}) - AIC: {model_fit.aic}")
                
                if model_fit.aic < best_aic:
                    best_aic = model_fit.aic
                    best_order = (p, d, q)
                    best_model = model_fit
            except:
                continue

print(f"Best ARIMA Order: {best_order}")
print(best_model.summary())

forecast_steps = 10
forecast = best_model.forecast(steps=forecast_steps)
forecast_index = range(2024, 2024 + forecast_steps)

plt.figure(figsize=(10, 6))
plt.plot(tc_loss_data, label='Observed')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.title('Tree Cover Loss Forecast')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np
plt.style.use('ggplot')

forest_df = pd.read_csv('./clean_data/forest.csv')
tc_loss_columns = [col for col in forest_df.columns if 'tc_loss_ha_' in col]
tc_loss_data = forest_df[tc_loss_columns].mean(axis=0)
tc_loss_data.index = range(2001, 2024)

tc_loss_data.plot(figsize=(10, 6))
plt.title('Tree Cover Loss (Ha) Over Time')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.show()

result = adfuller(tc_loss_data)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] > 0.05:
    print("The time series is non-stationary. Differencing is required.")
    tc_loss_data_diff = tc_loss_data.diff().dropna()
else:
    print("The time series is stationary.")
    tc_loss_data_diff = tc_loss_data

plot_acf(tc_loss_data_diff)
plt.show()
plot_pacf(tc_loss_data_diff)
plt.show()

best_aic = np.inf
best_order = None
best_seasonal_order = None
best_model = None

for p in range(0, 3):
    for d in range(0, 2):
        for q in range(0, 3):
            for P in range(0, 2):
                for D in range(0, 2):
                    for Q in range(0, 2):
                        try:
                            model = SARIMAX(tc_loss_data, order=(p, d, q), 
                                            seasonal_order=(P, D, Q, 12),
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
                            model_fit = model.fit(disp=False)
                            if model_fit.aic < best_aic:
                                best_aic = model_fit.aic
                                best_order = (p, d, q)
                                best_seasonal_order = (P, D, Q, 12)
                                best_model = model_fit
                        except:
                            continue

print(f"Best SARIMA Order: {best_order}")
print(f"Best Seasonal Order: {best_seasonal_order}")
print(best_model.summary())

forecast_steps = 10
forecast = best_model.forecast(steps=forecast_steps)
forecast_index = range(2024, 2024 + forecast_steps)

plt.figure(figsize=(10, 6))
plt.plot(tc_loss_data, label='Observed')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.title('Tree Cover Loss Forecast')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.style.use('ggplot')
import numpy as np

forest_df = pd.read_csv('./clean_data/forest.csv')
tc_loss_columns = [col for col in forest_df.columns if 'tc_loss_ha_' in col]
tc_loss_data = forest_df[tc_loss_columns].mean(axis=0)
tc_loss_data.index = range(2001, 2024)

plt.figure(figsize=(12, 6))
plt.plot(tc_loss_data.index, tc_loss_data, marker='o', linestyle='-', color='blue', label='Observed')
plt.title('Tree Cover Loss (Ha) Over Time')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.xticks(tc_loss_data.index, rotation=45)
plt.grid(True)
plt.legend()
plt.show()

result = adfuller(tc_loss_data)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] > 0.05:
    print("The time series is non-stationary. Differencing is required.")
    tc_loss_data_diff = tc_loss_data.diff().dropna()
else:
    print("The time series is stationary.")
    tc_loss_data_diff = tc_loss_data

plot_acf(tc_loss_data_diff)
plt.show()
plot_pacf(tc_loss_data_diff)
plt.show()

best_aic = np.inf
best_order = None
best_seasonal_order = None
best_model = None

for p in range(0, 3):
    for d in range(0, 2):
        for q in range(0, 3):
            for P in range(0, 2):
                for D in range(0, 2):
                    for Q in range(0, 2):
                        try:
                            model = SARIMAX(tc_loss_data, order=(p, d, q), 
                                            seasonal_order=(P, D, Q, 12),
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
                            model_fit = model.fit(disp=False)
                            if model_fit.aic < best_aic:
                                best_aic = model_fit.aic
                                best_order = (p, d, q)
                                best_seasonal_order = (P, D, Q, 12)
                                best_model = model_fit
                        except:
                            continue

print(f"Best SARIMA Order: {best_order}")
print(f"Best Seasonal Order: {best_seasonal_order}")
print(best_model.summary())

forecast_steps = 10
forecast = best_model.forecast(steps=forecast_steps)
forecast_index = range(2024, 2024 + forecast_steps)

plt.figure(figsize=(12, 6))
plt.plot(tc_loss_data.index, tc_loss_data, marker='o', linestyle='-', color='blue', label='Observed')
plt.plot(forecast_index, forecast, marker='s', linestyle='--', color='red', label='Forecast')
plt.title('Tree Cover Loss Forecast')
plt.xlabel('Year')
plt.ylabel('Tree Cover Loss (Ha)')
plt.xticks(list(tc_loss_data.index) + list(forecast_index), rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```
