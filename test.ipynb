```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import Input

def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data from {file_path}: {e}")
        return None

def prepare_train_test_data(district_data, sequence_length=10):
    try:
        losses = district_data.loc[:, 'tc_loss_ha_2001':'tc_loss_ha_2023'].values.flatten()
        if len(losses) <= sequence_length:
            print(f"Not enough data for sequence preparation in district: {district_data['subnational2'].iloc[0]}")
            return None, None, None, None
        
        scaler = MinMaxScaler()
        losses_scaled = scaler.fit_transform(losses.reshape(-1, 1)).flatten()
        
        X, y = [], []
        for i in range(len(losses_scaled) - sequence_length):
            X.append(losses_scaled[i:i + sequence_length])
            y.append(losses_scaled[i + sequence_length])
        
        X, y = np.array(X).reshape((-1, sequence_length, 1)), np.array(y)
        
        # Split into train/test (2001-2018 for training, 2019-2023 for testing)
        train_size = 18  # 2001â€“2018
        X_train, y_train = X[:train_size], y[:train_size]
        test_scaled = losses_scaled[train_size:]
        
        return X_train, y_train, test_scaled, scaler
    except Exception as e:
        print(f"Error preparing data: {e}")
        return None, None, None, None

def build_lstm_model(input_shape):
    try:
        model = Sequential([
            Input(shape=input_shape),
            LSTM(50, activation='relu'),
            Dropout(0.2),
            Dense(1)
        ])
        model.compile(optimizer='adam', loss='mean_squared_error')
        return model
    except Exception as e:
        print(f"Error building LSTM model: {e}")
        return None

def evaluate_predictions(test_scaled, predictions, scaler):
    try:
        test_actual = scaler.inverse_transform(test_scaled.reshape(-1, 1)).flatten()
        test_predicted = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()
        mae = mean_absolute_error(test_actual, test_predicted)
        mse = mean_squared_error(test_actual, test_predicted)
        r2 = r2_score(test_actual, test_predicted)
        return mae, mse, r2, test_actual, test_predicted
    except Exception as e:
        print(f"Error evaluating predictions: {e}")
        return None, None, None, None, None

def plot_all_years(district_name, train_years, train_values, test_years, test_actual, pred_years, predictions):
    train_values = train_values[-len(train_years):]  # Align values to match years if necessary
    plt.figure(figsize=(12, 6))
    plt.plot(train_years, train_values, label='Training Data (2001-2018)', color='blue', marker='o')
    plt.plot(test_years, test_actual, label='Actual Data (2019-2023)', color='green', marker='o')
    plt.plot(pred_years, predictions, label='Predictions (2019-2023)', color='red', linestyle='dashed', marker='o')
    plt.title(f"Prediction Results for {district_name}")
    plt.xlabel("Year")
    plt.ylabel("Forest Loss")
    plt.legend()
    plt.grid()
    plt.savefig(f"./plots/all_years_{district_name}.png")

def process_district(district_data):
    try:
        district_name = district_data['subnational2'].iloc[0]
        print(district_name)
        X_train, y_train, test_scaled, scaler = prepare_train_test_data(district_data)
        if X_train is None or y_train is None:
            return None
        
        model = build_lstm_model((X_train.shape[1], 1))
        model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0)
        
        # Predict future values
        predictions = []
        last_sequence = X_train[-1].reshape(1, X_train.shape[1], 1)
        for _ in range(len(test_scaled)):
            next_pred = model.predict(last_sequence, verbose=0)
            predictions.append(next_pred[0, 0])
            last_sequence = np.append(last_sequence[:, 1:, :], next_pred.reshape(1, 1, 1), axis=1)
        
        # Evaluate and plot results
        mae, mse, r2, test_actual, test_predicted = evaluate_predictions(test_scaled, predictions, scaler)
        print(f"District: {district_name} | MAE: {mae:.4f}, MSE: {mse:.4f}, R2: {r2:.4f}")
        
        train_years = list(range(2001, 2019))
        test_years = list(range(2019, 2024))
        pred_years = list(range(2019, 2024))
        train_values = scaler.inverse_transform(X_train[:, -1, 0].reshape(-1, 1)).flatten()
        plot_all_years(district_name, train_years, train_values, test_years, test_actual, pred_years, test_predicted)
        
        return test_predicted
    except Exception as e:
        print(f"Error processing district {district_name}: {e}")
        return None

def main():
    forest_path = './clean_data/forest.csv'
    forest_df = load_data(forest_path)
    if forest_df is None:
        return

    predictions_dict = {}
    districts = forest_df['subnational2'].unique()

    for district in districts:
        try:
            district_data = forest_df[forest_df['subnational2'] == district]
            #print(district_data)
            predictions = process_district(district_data)
            print(predictions)
            if predictions is not None:
                predictions_dict[district] = predictions
        except Exception as e:
            print(f"Error processing district {district}: {e}")
    
    predictions_df = pd.DataFrame(predictions_dict)
    predictions_df.to_csv('./clean_data/lstm_predictions.csv', index=False)
    print("Predictions saved successfully.")

if __name__ == "__main__":
    main()
```
